```python
# Identify numerical and categorical columns
num_cols = df.select_dtypes(include=['number']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Fill missing numerical values with median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing categorical values with mode (most common value)
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Verify missing values are handled
print("Missing values after cleaning:\n", df.isnull().sum())
```

```python
# check Percentage of missing values in each column
missing_values = df.isnull().sum()/len(df)*100
print("Missing values percentage in each column:\n",missing_values)
```

```python
# Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count} ")
```

```python
# Count the occurence of each violations type
violations_count = df['Violation_Type'].value_counts()

# Plot the top violations
plt.Figure(figsize=(10,5))
violations_count.plot(kind='bar', color='skyblue')
plt.title("Most common traffic violations")
plt.xlabel("Violation Type")
plt.ylabel('Number of Violations')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Count the number of violations per location
location_counts = df['Location'].value_counts()

# Plot the top locations
plt.figure(figsize=(10, 5))
location_counts.plot(kind="bar", color="orange")
plt.title("High Risk location for Traffic violations")
plt.xlabel("Location")
plt.ylabel('Number of violation')
plt.xticks(rotation = 45)
plt.show() 
```

```python
# Count the number of occurrences for each violation type
violation_counts = df['Violation_Type'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10,6))
plt.pie(violation_counts, labels=violation_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Distribution of Traffic Violation Types")
plt.show()
```

```python
# Count violations per location
location_counts = df['Location'].value_counts()

# Plot bar chart
plt.Figure(figsize=(12,6))
sns.barplot(x=location_counts.index, y=location_counts.values, palette='viridis')
plt.xticks(rotation = 45)
plt.xlabel("Location")
plt.ylabel("Number of Violations")
plt.title("Traffic Violations by Locations")
plt.show()
```

```python
# Define age bins
bins = [0, 18, 25, 35, 50, 100]
labels = ["Under 18", '18-25', '26-35', '36-50', 'Above 50']

# Create new Column "Age Group"
df['Age_Group'] = pd.cut(df['Driver_Age'], bins = bins, labels = labels, right = False)

# Count violations by Age Group
age_group_counts = df['Age_Group'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, hue=age_group_counts.index, palette='coolwarm', legend=False)
plt.xlabel("Age Group")
plt.ylabel("Number of Violations")
plt.title("Traffic Violations by Age Group")
plt.show()
```

```python
# Count of viollations by vehicle type
vehicle_type_counts = df['Vehicle_Type'].value_counts()

# Plot Pie Chart
plt.figure(figsize=(10,6))
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%', colors=sns.color_palette("Set2"))
plt.title("Distribution of Traffic Violations by Vehicle Type")
plt.show()
```
